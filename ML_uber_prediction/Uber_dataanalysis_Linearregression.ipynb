```python
import pandas as pd
data = pd.read_csv('taxi.csv')
data.head()
```

```python
x = data.drop('Numberofweeklyriders',axis =1)
y = data['Numberofweeklyriders']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, ytest = train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
print("the model score for trained adta",model.score(x_train,y_train))
```

```python
import pickle
pickle.dump(model,open('Uber_dataanalysis_Linearregression.pkl','wb'))
```

```python
model1 = pickle.load(open('Uber_dataanalysis_Linearregression.pkl','rb'))
print(model1.predict([[40,1695000,11600,100]]))
```
