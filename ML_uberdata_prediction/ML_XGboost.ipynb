```python
import pandas as pd
data = pd.read_csv('taxi.csv')
data.head()
```

```python
x= data.drop('Numberofweeklyriders',axis=1)
y= data['Numberofweeklyriders']
```

```python
from sklearn.model_selection import train_test_split
x_tain,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(
n_estimators = 200,
learning_rate=0.05,
max_depth=5,
random_state=42
)
model.fit(x_tain,y_train)
print('the model score of trained  data',model.score(x_tain,y_train))
```

```python
import pickle
pickle.dump(model,open("ML_xgboost.pkl",'wb'))
```

```python
model1 = pickle.load(open('ML_xgboost.pkl','rb'))
print('The predicted model score with xgboost',model1.predict([[65,1640000,12650,105]]))
```
